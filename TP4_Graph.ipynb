```python
import networkx as nx
import kagglehub
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Download latest version
path = kagglehub.dataset_download("mateuscco/toy-network-datasets")

print("Path to dataset files:", path)
```

```python
!ls {path}
```

## Q1. Chargement d'un graphe .net avec NetworkX


```python
G = nx.read_pajek(path+"/26KeroNetwork.net")

# Pajek charge un MultiGraph puis nous le convertissons en Graph simple
G = nx.Graph(G)
```

# Q2 : Affichage du nombre de nœuds et d’arêtes


```python
print("Nombre de nœuds :", G.number_of_nodes())
print("Nombre d’arêtes :", G.number_of_edges())
```

# Q3 : Visualisation le graphe avec matplotlib


```python
plt.figure(figsize=(8, 8))
nx.draw(G, with_labels=True, node_size=500, node_color="skyblue", edge_color="gray")
plt.title("Visualisation du graphe")
plt.show()
```

# Q4 : Calcul du degré de chaque nœud


```python
degrees = dict(G.degree())
print("Degrés des nœuds :")
degrees
```

# Q5 : Traçage de la distribution des degrés (histogramme)


```python
plt.figure(figsize=(8, 6))
plt.hist(list(degrees.values()), bins=np.arange(0, max(degrees.values())+2)-0.5,
         edgecolor='black', color="orange")
plt.title("Distribution des degrés")
plt.xlabel("Degré")
plt.ylabel("Nombre de nœuds")
plt.show()
```

# Q6 : Visualisation avec ipycytoscape


```python
from ipycytoscape import CytoscapeWidget

cyto = CytoscapeWidget()
cyto.graph.add_graph_from_networkx(G)
cyto.set_style([{
    "selector": 'node',
    "style": {"content": "data(id)", "background-color": "skyblue"}
}, {
    "selector": 'edge',
    "style": {"line-color": "gray"}
    }])
```
