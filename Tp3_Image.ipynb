```python
import os, random, kagglehub
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

## Q1 : Chargement d'un petit dataset d’images

```python
path = kagglehub.dataset_download("msarmi9/food101tiny")

print("Path to dataset files:", path)
print(path)
```

```python
data_path = os.path.join(path, 'data', 'food-101-tiny', 'train', 'apple_pie')
data_path
```

```python
data_dir = data_path

image_files = os.listdir(data_dir)

print('Number of images in folder : ', len(image_files))
print('First 5 images : ', image_files[:5])
```

## Q2 : Affichage de la dimension d’une image

```python
# Choisir une image au hasard
img_path = random.choice(image_files)
# Charger l'image avec PIL
img_path = os.path.join(data_dir, img_path)
image = Image.open(img_path)
print("Selected : ", os.path.basename(img_path))
print("Dimension de l'image :", image.size)  # (largeur, hauteur)
```

## Q3 : Visualisation de quelques images avec matplotlib

```python
plt.figure(figsize=(6,6))
for i in range(9):
    img = Image.open(os.path.join(data_dir, image_files[i]))
    plt.subplot(3,3,i+1)
    plt.imshow(img)
    plt.axis("off")
plt.show()
```

## Q4 : Normalisation des pixels (0–1)

```python
img_array = np.array(image) / 255.0
print("Pixels normalisés (0–1) :")
print(img_array[:2, :2, :])  # un petit extrait
```

## Q5 : Calcul de la moyenne et de l’écart-type des pixels d’une image

```python
mean = np.mean(img_array)
std = np.std(img_array)

print("Moyenne des pixels :", mean)
print("Écart-type des pixels :", std)
```
